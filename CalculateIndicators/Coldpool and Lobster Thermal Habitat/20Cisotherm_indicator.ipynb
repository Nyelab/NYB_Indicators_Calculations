```python
def inNYB(longitude, latitude, time, sst, shape):
    X = longitude
    Y = latitude
    y, x = np.meshgrid(Y,X)
    empty = np.empty([len(time), len(latitude), len(longitude)])
    empty[:,:,:] = np.nan
    for i in range(len(longitude)):
        for j in range(len(latitude)):
            if Point(x[i,j], y[i,j]).within(shape) == True:
                empty[:,j,i] = sst[:,i,j]
    return empty
```
